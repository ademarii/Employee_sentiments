### Employee Sentiment Analysis

#### Load the necessary libraries

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download the VADER lexicon for sentiment analysis
nltk.download('vader_lexicon')
```

#### Create functions
- Read file and
- Analyze text

```python
def analyze_sentiment(text):
    sia = SentimentIntensityAnalyzer()
    sentiment_dict = sia.polarity_scores(text)
    compound = sentiment_dict['compound']
    if compound >= 0.05:
        return 'Positive'
    elif compound <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

def read_feedback(file_path):
    with open(file_path, 'r') as file:
        feedback_list = file.readlines()
    return feedback_list

file_path = 'employee_feedback2.csv'
employee_feedback = read_feedback(file_path)

all_feedback = " ".join(feedback.strip() for feedback in employee_feedback)
for feedback in employee_feedback:
    sentiment = analyze_sentiment(feedback.strip())
    print(f"Feedback: {feedback.strip()}\nSentiment: {sentiment}\n")
```

```python
# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_feedback)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
